```
import pandas as pd
import numpy as np
import datetime
```

```
movies_data1 = pd.DataFrame ( movies_data)
movies_data1['Release_year'] = pd.DatetimeIndex(movies_data['Release_Date']).year
```

```
movies_data1.Major_Genre.unique()
```

```
sns.scatterplot(x='Release_year', y='US_Gross', data= movies_data1, hue= 'Major_Genre', size= 'Production_Budget', palette= 'Paired')
plt.ylabel('Gross Revenue in US (in hundred million $)')
for i, movie in enumerate(movies_data1.Title):
  if movies_data1.Release_year[i]>2010 and movies_data1.US_Gross[i] > 25000000 or movies_data1.US_Gross[i] > 450000000 or movies_data1.Release_year[i] < 1929 :
    plt.annotate( movie, (movies_data1.Release_year[i], movies_data1.US_Gross[i]))

plt.title("Movie's Gross Revenue Generation over time in US")
plt.show()
```

```
sns.scatterplot(x='Release_year', y='Worldwide_Gross', data= movies_data1, hue= 'Major_Genre' )

for i, movie in enumerate(movies_data1.Title):
  if movies_data1.Release_year[i]>2020 and movies_data1.Worldwide_Gross[i] > 50000000 or movies_data1.Worldwide_Gross[i] > 1000000000:
    plt.annotate( movie, (movies_data1.Release_year[i], movies_data1.Worldwide_Gross[i]))

plt.title("Movie's Gross Revenue(Worldwide) over time")
plt.show()
```
