```
plt.figure(figsize=(15,6))
barplot_title_loan_amt=sns.barplot(x='title', y='loan_amnt', data= lending_club_data)
barplot_title_loan_amt.set_xticklabels(barplot_title_loan_amt.get_xticklabels(), rotation=45)
plt.show()
```

```
plt.figure(figsize=(15,6))
catplot_title_loan_amt=sns.catplot(x='title', y='loan_amnt', data= lending_club_data)
catplot_title_loan_amt.set_xticklabels(rotation=90)
plt.show()
```

```
sns.distplot(lending_club_data.loan_amnt, rug=True)
plt.legend()
```

```
lending_club_data['loan_amnt'].equals(lending_club_data['funded_amnt'])
```

```
lending_club_data['funded_amnt_inv'].equals(lending_club_data['funded_amnt'])
```

```
sns.scatterplot(x='funded_amnt', y='funded_amnt_inv', data= lending_club_data)
plt.title('Variation between fund amount given and invested')
plt.show()
```

The data shows that there is hardly any variation between the fund granted and invested

```
sns.pairplot(lending_club_data, hue='title', x_vars='grade', y_vars='int_rate', kind='scatter', aspect=1, dropna=True)
```

```
g = sns.FacetGrid(lending_club_data, col="title", col_wrap=4, height=3)
g.map(sns.scatterplot, "grade", "int_rate", color=".3")
```

```
plt.figure(figsize=(15,6))
k=sns.scatterplot(x='grade', y='int_rate', data= lending_club_data, hue= 'title', palette= 'Paired')
plt.title('Intrest rate increases as the grading changes from A to G')
k.legend(loc='best', frameon=False,)
```

This Graph shows that A grade are charge with lower intrest rates and intrest rate increases as the Grading increase from A to g

```
sns.scatterplot('loan_amnt','term', data= lending_club_data)
plt.title('Variation of loan amount w.r.t terms')
plt.show()
```
